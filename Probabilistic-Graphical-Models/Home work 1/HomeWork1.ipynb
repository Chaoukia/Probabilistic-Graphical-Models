```python
import numpy as np
import matplotlib.pyplot as plt
from hw1_toolbox import *
%matplotlib inline
```

# Loading the data

```python
# Loading training data
train = np.genfromtxt('classificationA.train')
X_A = train[:, 0:2]
y_A = train[:, 2]

train = np.genfromtxt('classificationB.train')
X_B = train[:, 0:2]
y_B = train[:, 2]

train = np.genfromtxt('classificationC.train')
X_C = train[:, 0:2]
y_C = train[:, 2]

# Loading test data
test = np.genfromtxt('classificationA.test')
X_A_test = test[:, 0:2]
y_A_test = test[:, 2]

test = np.genfromtxt('classificationB.test')
X_B_test = test[:, 0:2]
y_B_test = test[:, 2]

test = np.genfromtxt('classificationC.test')
X_C_test = test[:, 0:2]
y_C_test = test[:, 2]
```

```python
def plots(clas):
    """
    Description
    -------------
    This function plots the training and test data points along with the line p(y=1|x) = 0.5 for every model.
    
    Parameters
    -------------
    clas : a class, LDA, QDA, LogisticRegression or LinearRegression.
    """
    plt.figure(figsize = (20, 10))

    plt.subplot(2, 3, 1)
    model = clas()
    model.fit(X_A, y_A)
    model.plot_line(X_A, y_A, title = "Training A : error = %.4f" %error(y_A, model.predict(X_A)))

    plt.subplot(2, 3, 2)
    model = clas()
    model.fit(X_B, y_B)
    model.plot_line(X_B, y_B, title = "Training B : error = %.4f" %error(y_B, model.predict(X_B)))

    plt.subplot(2, 3, 3)
    model = clas()
    model.fit(X_C, y_C)
    model.plot_line(X_C, y_C, title = "Training C : error = %.4f" %error(y_C, model.predict(X_C)))

    plt.subplot(2, 3, 4)
    model = clas()
    model.fit(X_A, y_A)
    model.plot_line(X_A_test, y_A_test, title = "Test A : error = %.4f" %error(y_A_test, model.predict(X_A_test)))

    plt.subplot(2, 3, 5)
    model = clas()
    model.fit(X_B, y_B)
    model.plot_line(X_B_test, y_B_test, title = "Test B : error = %.4f" %error(y_B_test, model.predict(X_B_test)))

    plt.subplot(2, 3, 6)
    model = clas()
    model.fit(X_C, y_C)
    model.plot_line(X_C_test, y_C_test, title = "Test C : error = %.4f" %error(y_C_test, model.predict(X_C_test)))

    plt.show()
```

# Generative model LDA

```python
plots(LDA)
```

# Logistic Regression

```python
plots(LogisticRegression)
```

# Linear Regression

```python
plots(LinearRegression)
```

# QDA

```python
plots(QDA)
```
